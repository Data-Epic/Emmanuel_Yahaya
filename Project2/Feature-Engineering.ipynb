```python
worksheet = pd.read_html('https://docs.google.com/spreadsheets/d/1SUlcukpgxf6pfFQbj6DKQZXSlOhyBOSuBdbmQ2ZbhTM/edit#gid=327494290', header=1)

worksheet = worksheet [0]
```

```python
worksheet.isnull().sum()
```

```python
worksheet_1 = worksheet.drop(["Unnamed: 44", "Unnamed: 45", "TIDF Compliance", "Site Permit", "Street Number Suffix", "Unit Suffix", "Completed Date", "Structural Notification", "Voluntary Soft-Story Retrofit", "Fire Only Permit"], axis=1)
```

```python
pd.set_option('display.max_columns', None)
worksheet_1
```

###### DATETIME FEATURE

```python
worksheet_1['Permit Creation Date'] = pd.to_datetime(worksheet_1['Permit Creation Date'])

# Extract year, month, day, and day of the week as new features
worksheet_1['Year'] = worksheet_1['Permit Creation Date'].dt.year
worksheet_1['Month name'] = worksheet_1['Permit Creation Date'].dt.month_name()
worksheet_1['Day name'] = worksheet_1['Permit Creation Date'].dt.day_name()
worksheet_1['Day of Week'] = worksheet_1['Permit Creation Date'].dt.dayofweek  # Monday=0, Sunday=6

print(worksheet_1.head(99))
```

```python
worksheet_1.head()
```

###### FEATURE SHOWING PERCENTAGE OF MISSING VALUES

```python
total_columns = len(worksheet_1.columns)
```

```python
total_columns
```

```python
missing_values_count = worksheet.isnull().sum()

worksheet_1['Missing Values Percentage'] = (missing_values_count / total_columns) * 100
```

```python
worksheet_1.head()
```

###### FEATURE SHOWING DURATION

```python
worksheet_1["Issued Date"]
```

```python
IssuedDate_1 = worksheet_1['Issued Date'].head(99).fillna(method = 'bfill')
```

```python
worksheet_1['Issued Date'] = worksheet_1['Issued Date'].fillna(IssuedDate_1)
```

```python
worksheet_1["Issued Date"]
```

```python
worksheet_1['Permit Creation Date'] = pd.to_datetime(worksheet_1['Permit Creation Date'])
worksheet_1['Issued Date'] = pd.to_datetime(worksheet_1['Issued Date'])
```

```python
worksheet_1['Duration (days)'] = (worksheet_1['Issued Date'] - worksheet_1['Permit Creation Date']).dt.days
```

```python
worksheet_1.head()
```

###### CATEGORICAL ENCODING USING ONE HOT ENCODING

```python
columns=["Permit Type Definition", "Current Status", "Existing Use"]
dummyData = pd.get_dummies(worksheet_1[columns])
dummyData
```

```python
worksheet_1 = pd.concat([worksheet_1, dummyData], axis=1)
worksheet_1.head()
```

```python
worksheet_1.to_csv('Revised_Data.csv', index=False)
```
